```python
from transformers import VitsModel, AutoTokenizer
import torch

model = VitsModel.from_pretrained("facebook/mms-tts-kzf")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-kzf")

text = "Assalomu aleykum Malika opa yaxshimisiz"
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform
```

```python
import scipy
import numpy as np

# Convert the PyTorch tensor to a NumPy array
output_numpy = output.detach().cpu().numpy()

# Convert the floating-point audio data to 16-bit integers
# Scale the data to the range of a 16-bit integer
output_int16 = np.int16(output_numpy * 32767)

scipy.io.wavfile.write
```
